```python

# okay can you come up with a new thing
# you want to write all claims! but with references working properly

# print(len(graph))
from wikibaseintegrator import wbi_core, wbi_login

# login to the thing
# login_instance = wbi_login.Login(user='paulduchesne', pwd='petecrenshaw', mediawiki_api_url='http://localhost:8181/w/api.php')


# okay you need to write things 

elements = pydash.uniq([a for a,b,c in graph.triples((None, wb_n['#claim'], None))])
# print(len(elements))

start = datetime.datetime.now()

for numb, elem in enumerate(elements):
    numb = numb+1
    
    #     point = (row.name+1)
    duration = (datetime.datetime.now()-start)/numb # this is how long per loop
    eta = (duration * (len(elements)-numb))+datetime.datetime.now()
    print('running:', numb, 'of', len(elements), ';', round((numb/len(elements))*100, 2), '% ; eta:', eta)
    clear_output(wait=True)    
    

    
#     claimer = list()

    collected = [(a,b,c) for a,b,c in graph.triples((elem, wb_n['#claim'], None))]

    new_list = list()
    
    for a,b,c in collected:
        for d,e,f in graph.triples((c, None, None)):
            if e != wb_n['#reference'] and e != wb_n['#qualifier']:
                prop_order = str(e.replace('P','')).zfill(4)
                new_list.append((c,e, prop_order))
                

    new_list = sorted(new_list, key=lambda x:x[2])
    
    for y in sorted(pydash.uniq([x[2] for x in new_list])):
        sub_list = [x for x in new_list if x[2] == y]

        claimer = list()    


        for c, n, xx in sub_list:
            
            quals = list()
            for d,e,f in graph.triples((c, None, None)):
                if e == wb_n['#qualifier']:
                    quals.append(f)

            qual2 = list()
            for q in quals:
                for g,h,i in graph.triples((q, None, None)):

                    if isinstance(i, rdflib.URIRef):                    
           
                        qual2.append(wbi_core.ItemID(value=i, prop_nr=h, is_qualifier=True))
                    elif isinstance(i, rdflib.Literal):            
                        qual2.append(wbi_core.String(value=i, prop_nr=h, is_qualifier=True)) 
                    else:
                        raise Exception('no proper datatype!')                          
            
            
#                     print(h,i)

            refs = list()
            for d,e,f in graph.triples((c, None, None)):
                if e == wb_n['#reference']:
                    refs.append(f)

            refs2 = list()
            for r in refs:
                for g,h,i in graph.triples((r, None, None)):
                    refs2.append([wbi_core.ItemID(value=i, prop_nr=h, is_reference=True)])

            for d,e,f in graph.triples((c, None, None)):
                if e != wb_n['#reference'] and e != wb_n['#qualifier']:
                    if isinstance(f, rdflib.URIRef):
                        claimer.append(wbi_core.ItemID(value=str(f), prop_nr=str(e), qualifiers=qual2, references=pydash.uniq(refs2)))
                    elif isinstance(f, rdflib.Literal):
                        claimer.append(wbi_core.String(value=str(f), prop_nr=str(e), qualifiers=qual2, references=pydash.uniq(refs2)))
                    else:
                        raise Exception('no proper datatype!')  

        wd_item = wbi_core.ItemEngine(item_id=str(elem), data=claimer, mediawiki_api_url='http://localhost:8181/w/api.php')
        wd_item.write(login_instance)  

print('all done.')
```
