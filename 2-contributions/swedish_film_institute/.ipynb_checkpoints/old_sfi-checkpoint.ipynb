```python

# define graph and namespaces

graph = rdflib.Graph()

name_sfi = rdflib.Namespace('https://www.filminstitutet.se/') 
name_wb = rdflib.Namespace('http://wikibas.se/ontology')
name_fiaf = rdflib.Namespace("https://www.fiafnet.org/")
```

```python

# institute specific entities

graph.add((name_sfi['ontology/property/sfi_work_id'], rdflib.RDFS.label, rdflib.Literal('Swedish Film Institute work ID'))) 
graph.add((name_sfi['ontology/property/sfi_agent_id'], rdflib.RDFS.label, rdflib.Literal('Swedish Film Institute agent ID'))) 

claim1 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
graph.add((name_sfi['ontology/item/sfi'], rdflib.RDFS.label, rdflib.Literal('Swedish Film Institute', lang='en'))) 
graph.add((name_sfi['ontology/item/sfi'], rdflib.RDFS.label, rdflib.Literal('Svenska Filminstitutet', lang='sv'))) 
graph.add((name_sfi['ontology/item/sfi'], name_wb['#claim'], claim1))
graph.add((claim1, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/holding_institution'])) 

claim2 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
graph.add((name_sfi['ontology/item/originaltitel'], rdflib.RDFS.label, rdflib.Literal('original title', lang='en'))) 
graph.add((name_sfi['ontology/item/originaltitel'], rdflib.RDFS.label, rdflib.Literal('originaltitel', lang='sv'))) 
graph.add((name_sfi['ontology/item/originaltitel'], name_wb['#claim'], claim2)) 
graph.add((claim2, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/title_type'])) 

claim3 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
graph.add((name_sfi['ontology/item/roll'], rdflib.RDFS.label, rdflib.Literal('roll', lang='sv'))) 
graph.add((name_sfi['ontology/item/roll'], name_wb['#claim'], claim3)) 
graph.add((claim3, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/agent_type'])) 

print(len(graph))
```

```python

# write work specifics

for x in data:
    work_id = x['object_number']
    work = name_sfi[f"resource/work/{work_id}"]
    
    claim1 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
    claim2 = name_sfi[f"resource/claim/{uuid.uuid4()}"]    

    graph.add((work, name_wb['#claim'], claim1))
    graph.add((claim1, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/work']))

    graph.add((work, name_wb['#claim'], claim2))
    graph.add((claim2, name_sfi['ontology/property/sfi_work_id'], rdflib.Literal(work_id)))    

print(len(graph))
```

```python

# write original title

def title_array(data):
    array = list()
    if pydash.get(x, 'Title.title_complete'):
        array.append(pydash.get(x, 'Title'))
    else:
        for y in pydash.get(x, 'Title'):
            array.append(y)  
    return array

for x in data:
    
    work_id = x['object_number']
    work = name_sfi[f"resource/work/{work_id}"]    
    
    for t in title_array(x):
        if 'Originaltitel' in str(t):

            claim1 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
            qual1 = name_sfi[f"resource/qualifier/{uuid.uuid4()}"]        
            ref1 = name_sfi[f"resource/reference/{uuid.uuid4()}"] 
 
            graph.add((work, name_wb['#claim'], claim1))
            graph.add((claim1, name_fiaf['ontology/property/title'], rdflib.Literal(t['title_complete'])))
        
            graph.add((claim1, name_wb['#qualifier'], qual1))
            graph.add((qual1, name_fiaf['ontology/property/title_type'], name_sfi['ontology/item/originaltitel']))
            
            graph.add((claim1, name_wb['#reference'], ref1))
            graph.add((ref1, name_fiaf['ontology/property/contributed_by'], name_sfi['ontology/item/sfi']))  
    
print(len(graph))  
   
```

```python

# write cast agents


def reference(claim_id):
    
    ref_id = name_sfi[f"resource/reference/{uuid.uuid4()}"]
    graph.add((claim_id, name_wb['#reference'], ref_id))
    graph.add((ref_id, name_fiaf['ontology/property/contributed_by'], name_sfi['ontology/item/sfi']))  
    

for x in data:
    if 'Cast' in x:
        cast = pydash.get(x, 'Cast')
        for c in cast:
            
            work = name_sfi[f"resource/work/{x['object_number']}"]            
            agent1 = name_sfi[f"resource/agent/{c['cast.name']['id_number']}"]
            
            claim1 = name_sfi[f"resource/claim/{uuid.uuid4()}"]
            claim2 = name_sfi[f"resource/claim/{uuid.uuid4()}"]    
            claim3 = name_sfi[f"resource/claim/{uuid.uuid4()}"]   
            claim4 = name_sfi[f"resource/claim/{uuid.uuid4()}"]  
            claim5 = name_sfi[f"resource/claim/{uuid.uuid4()}"]  
            claim6 = name_sfi[f"resource/claim/{uuid.uuid4()}"]              
            qual1 = name_sfi[f"resource/qualifier/{uuid.uuid4()}"]  
            qual2 = name_sfi[f"resource/qualifier/{uuid.uuid4()}"]   
            role1 = name_sfi[f"resource/role/{uuid.uuid4()}"]               
            
            graph.add((work, name_wb['#claim'], claim1))
            graph.add((claim1, name_fiaf['ontology/property/agent_cast'], agent1))
            
            graph.add((claim1, name_wb['#qualifier'], qual1))
            graph.add((qual1, name_fiaf['ontology/property/agent_type'], name_sfi['ontology/item/roll'])) # agent_type

#             graph.add((claim1, name_wb['#qualifier'], qual2))
#             graph.add((qual2, name_fiaf['ontology/property/agent_type'], name_sfi['ontology/item/roll'])) # agent_type
            
            
            graph.add((claim1, name_wb['#qualifier'], qual2))
            graph.add((qual2, name_fiaf['ontology/property/role'], role1))        

            graph.add((role1, name_wb['#claim'], claim6))
            graph.add((claim6, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/role']))      
                
            
            graph.add((role1, rdflib.RDFS.label, rdflib.Literal(c['cast.credit_on_screen'], lang='sv'))) # instane of title type (item)
                
            
            
#             credit_on_screen
            
            
            
            reference(claim1)
            
            graph.add((agent1, name_wb['#claim'], claim2))
            graph.add((claim2, name_fiaf['ontology/property/instance_of'], name_fiaf['ontology/item/agent']))
            
            label_name = ''
            
            if 'forename' in c['cast.name']:
                label_name += c['cast.name']['forename']
                graph.add((agent1, name_wb['#claim'], claim3))
                graph.add((claim3, name_fiaf['ontology/property/forename'], rdflib.Literal(c['cast.name']['forename'])))
                
                reference(claim3)

            graph.add((agent1, name_wb['#claim'], claim4))
            graph.add((claim4, name_fiaf['ontology/property/surname'], rdflib.Literal(c['cast.name']['surname'])))
            label_name += ' '+c['cast.name']['surname']
            reference(claim4)            
              
            graph.add((agent1, name_wb['#claim'], claim5))
            graph.add((claim5, name_sfi['ontology/property/sfi_agent_id'], rdflib.Literal(c['cast.name']['id_number'])))            
            
            
#             agent1
            
            graph.add((agent1, rdflib.RDFS.label, rdflib.Literal(label_name.strip(), lang='sv'))) 
        
        
        
        
        

# YU NEED TO ADD SWEDISH ROLE HERE            
            
            
print(len(graph))
```
